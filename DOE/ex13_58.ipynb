```python
std2 = df[1].values.std()
mean2 = mean[0]
t = stats.t.ppf(1-alpha/2, n, a)
xmin = mean2 - t*std2/(np.sqrt(n))
xmax = mean2 + t*std2/(np.sqrt(n))

mu = np.array([xmin, xmax])

print(mu)
```
