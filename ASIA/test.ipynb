```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh_tridiagonal
```

```python
L = 5e-2
m = 9.1093837015e-31
hbar = 1.05457182e-34
```

```python
N = 1000
x = np.linspace(0, L, N+1)
dx = L/N

V_x = 0*x
```

```python
main = 1/dx**2 + V_x[1:-1]
off = -1/(2*dx**2) * np.ones(len(main)-1)
E, psi = eigh_tridiagonal(main, off)
psi = psi.T
```

```python
E[0]*m*L**2/(hbar)**2
```
