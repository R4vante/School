## Question
According to the special theory of relativity, the factor $\gamma$ that determines the length contraction and the time dilation is given by

$$\gamma\ =\ \frac{1}{\sqrt{1 - \frac{v^2}{c^2}}}$$

Determine the numerical values of $\gamma$ for an object moving at speed $v = 0.01c$, $0.05c$, $0.10c$, $0.20c$, $0.30c$, $0.40c$, $0.50c$, $0.60c$, $0.70c$, $0.80c$, $0.90c$, and $0.99c$. Make a graph of $\gamma$ versus $v$.


## answer
To calculate the numeriacal values, first a function is created for $\gamma$

```python
def gamma(v):
    c = 1 # lightspeed is scaled to 1 because v in terms of c
    gamma = 1/(np.sqrt(1-(v**2/c**2)))
    return gamma
```

Now, create a vector containing the values for v.

```python
v_cal = [0.01, 0.05, 0.10, 0.20, 0.30, 0.40, 0.50, 0.60, 0.70, 0.80, 0.90, 0.99] # this is v/c
```

Now calculate gamma and print the result for every speed

```python
for i in v_cal:
    gam = gamma(i)
    print('for v = %.2f, gamma = %.2f' %(i, gam))
```

To create a figure, just create a vector containing different speeds (in v/c) and calculate gamma. In this case $v$ cannot be equal to $c$, otherwise python will complain. Afterwards, create the graph using matplotlib.

```python
v = np.linspace(0, 0.999, 1000)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(v, gamma(v))
ax.set_xlabel(r'$\frac{v}{c}$', fontsize=20)
ax.set_ylabel(r'$\gamma$', fontsize=20)
ax.set_title('Lorentz factor', fontsize=30)
```
